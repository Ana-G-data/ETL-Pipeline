```python
import requests
import re
from bs4 import BeautifulSoup
url_paises = 'https://statisticstimes.com/geography/countries-by-continents.php'
html_paises = requests.get(url_paises)
```

```python
html_paises.status_code
```

```python
sopa_paises = BeautifulSoup(html_paises.content, "html.parser")
```

```python
paises = sopa_paises.find_all("table")
len(paises)
#paises[2]
```

```python
lista_paises = paises[2]
```

```python
# con esto sacamos todas las filas de nuestra tabla

len(lista_paises.find_all("tr"))
```

```python
paises_dict = {
    'Country':[],
    'ISO_Code':[],
    'M49_Code':[],
    'Region1':[],
    'Region2':[],
    'Continent':[]
}
```

```python
for f in lista_paises.find_all('tr')[1:]:
    columna = [elemento.text for elemento in f.find_all('td')]

    print(columna)
```

```python
keys = paises_dict.keys()
for f in lista_paises.find_all('tr')[1:]:
    columna = [elemento.text for elemento in f.find_all('td')]
    
    for num, key in zip(range(1, len(columna)), keys):
        #print(key, num)
        paises_dict[key].append(columna[num])
```

```python
paises_dict
```

```python
df_paises = pd.DataFrame(paises_dict)
df_paises.head()
```

```python
df_paises.to_csv("datos/paises.csv",index=False, header=True)
```
