```python
import requests
import re
from bs4 import BeautifulSoup
url_paises = 'https://statisticstimes.com/geography/countries-by-continents.php'
html_paises = requests.get(url_paises)
```

```python
sopa_paises = BeautifulSoup(html_paises.content, "html.parser")
```

```python
paises = sopa_paises.find_all("table")
len(paises)
#paises[2]
```

```python
lista_paises = paises[2]
```

```python
# con esto sacamos todas las filas de nuestra tabla

len(lista_paises.find_all("tr"))
```

```python
paises_dict = {
    'Country':[],
    'ISO_Code':[],
    'M49_Code':[],
    'Region1':[],
    'Region2':[],
    'Continent':[]
}
```

```python
for f in lista_paises.find_all('tr')[1:]:
    columna = [elemento.text for elemento in f.find_all('td')]

    print(columna)
```

```python
keys = paises_dict.keys()
for f in lista_paises.find_all('tr')[1:]:
    columna = [elemento.text for elemento in f.find_all('td')]
    
    for num, key in zip(range(1, len(columna)), keys):
        
        paises_dict[key].append(columna[num])
```

```python
with open("../datos/arrivals_data.pickle", "rb") as f:
    data_flight = pickle.load(f)
```

```python
flight_data_list = []
for data in data_flight[0]['data']:
    flight_data = {}
    if 'terminal' in data['arrival']:
        flight_data["terminal"] = data['arrival']['terminal']
    else:
        flight_data["terminal"] = "No se ha encontrado la terminal de llegada."
    if 'scheduledTime' in data['arrival']:
        flight_data["scheduledTime"] = data['arrival']['scheduledTime']
    else:
        flight_data["scheduledTime"] = "No se ha encontrado el horario programado."
    if 'name' in data['airline']:
        flight_data["airlineName"] = data['airline']['name']
    else:
        flight_data["airlineName"] = "No se ha encontrado el nombre de la aerolínea."
    if 'iataCode' in data['airline']:
        flight_data["iataCode"] = data['airline']['iataCode']
    else:
        flight_data["iataCode"] = "No se ha encontrado el código IATA de la aerolínea."
    if 'number' in data['flight']:
        flight_data["flightNumber"] = data['flight']['number']
    else:
        flight_data["flightNumber"] = "No se ha encontrado el número del vuelo."
    if 'iataNumber' in data['flight']:
        flight_data["iataNumber"] = data['flight']['iataNumber']
    else:
        flight_data["iataNumber"] = "No se ha encontrado el número IATA del vuelo."
    flight_data_list.append(flight_data)
flight_data_df = pd.DataFrame(flight_data_list)
flight_data_df

```

```python
#Definimos una funcion para separar el scheduledTime en scheduledDay y la hora y luego lo applicamos a la columna
def split_scheduled_time(scheduled_time):
    if scheduled_time == "No se ha encontrado el horario programado.":
        return ("No se pudo separar", "No se ha encontrado el horario programado")
    else:
        scheduled_day, scheduled_time = scheduled_time.split("t")
        return scheduled_day, scheduled_time

flight_data_df["scheduledDay"], flight_data_df["scheduledTime"] = zip(*flight_data_df["scheduledTime"].apply(split_scheduled_time))
```
